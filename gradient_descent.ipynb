```python
import numpy as np

bias=0.0
weight=0.1
alpha=0.01
epoch=1000

def compute_cost(x,y,bias,weight):
    m=len(y)
    predictions=x*weight + bias
    return (1/(2*m))*np.sum(predictions-y)**2

def gradient_descent(x,y,bias,weight,learning_rate_alpha,epochs):
    m=x.shape[0]
    cost_history=[]

    for epochs in range(epochs):
        predictions=x*weight + bias
        error=predictions-y

        bias=bias - learning_rate_alpha * ((1/m)*np.sum(error))
        weight=weight - learning_rate_alpha * ((1/m)*np.sum(error*x))

        cost=compute_cost(x,y,bias,weight)
        cost_history.append(cost)

        if epoch%100 == 0:
            print(f"Epoch: {epoch}, Cost: {cost}")
            
    return bias,weight,cost_history


bias,weight,compute_cost=gradient_descent(x,y,bias,weight,alpha,epoch)
print(f"Optimized bias = {bias}, weight = {weight}, compute_cost = {compute_cost}")


      
```

# Stochastic Gradient Descent


```python
import numpy as np
def mean_squared_error(y,y_cap):
    return np.mean((y-y_cap)**2)


def stochastic_gradient_descent(X,y,leanring_rate=0.01,epochs=1000):
    n_samples=X.shape[0]
    n_features=X.shape[1]
    bias=0.1
    weights=np.zeros(n_features,dtype=np.float16)

    for epoch in range(epochs):
        random_index=np.random.randint(n_samples)
        xi=x[random_index:random_index+1]
        yi=y[random_index]

        ypred=np.dot(xi,weights)+bias
        weights=weights-learning_rate*(-2*xi.T.dot(yi-ypred))
        bias=bias-leanring_rate*(-2*(yi-ypred))

        if epoch%100==0:
            mse=mean_squared_error(y,X.dot(weights))
            print(f"Epoch {epoch}: mean_squared_error = {np.round(mse,2)}")
    return weights, bias


x=np.random.randint(1,5,size=(4,2))
y=np.random.randint(1,10,size=(4,1))

x=(x-np.mean(x,axis=0))/np.std(x,axis=0)
learning_rate=0.01
epochs=10000
weights, bias=stochastic_gradient_descent(x,y)

print(f"Final weight {weights}")
print(f"Final bias {bias}")

```

# Mini batch Gradient Descent


```python
def mse(y,ycap):
    return np.mean((y-ycap)**2)

def mini_batch_grad_descent(x,y,learning_rate=0.01,epochs=1000,batch_size=16):
    n_samples,n_features = x.shape
    bias=0.1
    weights=np.zeros(n_features,dtype=np.float16)

    for epoch in range(epochs):
        for i in range(0,n_samples,batch_size):
            x_batch=x[i:i+batch_size]
            y_batch=y[i:i+batch_size]

            y_pred_batch=np.dot(x_batch,weights)+bias

            weights=weights-((-2/batch_size) * np.dot(x_batch.T,(y_batch-y_pred_batch)))*learning_rate
            bias=bias-(learning_rate* ((2/batch_size) * np.sum(y_batch-y_pred_batch)))

        if epoch%100==0:
                y_cap=np.dot(x,weights)+bias
                mse_value=mse(y,y_cap)
                print(f"Epoch {epoch}: mean squared error {mse_value}")
                print(f"Weight: {weights}")
                print(f"Bias: {bias}")
    return weights,bias

x = np.random.randint(1, 10, size=(8, 2))
y = np.random.randint(1, 10, size=(8, ))

x = (x - np.mean(x, axis=0)) / np.std(x, axis=0)
weights,bias=mini_batch_grad_descent(x,y)
print("Final weights: ",weights)
print("Final bias: ",bias)
```

# RMS prop optimization

```python
import numpy as np

def mse(y, y_cap):
    return np.mean((y - y_cap) ** 2)

def mini_batch_grad_descent(x, y, learning_rate=0.01, epochs=1000, batch_size=16):
    n_samples, n_features = x.shape
    bias = 0.1
    weights = np.zeros(n_features)

    for epoch in range(epochs):
        for i in range(0, n_samples, batch_size):
            x_batch = x[i:i + batch_size]
            y_batch = y[i:i + batch_size]

            y_pred_batch = np.dot(x_batch, weights) + bias  

            y_batch = y_batch.reshape(-1, 1)                
            y_pred_batch = y_pred_batch.reshape(-1, 1)     

            error = (y_batch - y_pred_batch).flatten()       
            gradient_w = (-2 / batch_size) * np.dot(x_batch.T, error)  
            gradient_b = (-2 / batch_size) * np.sum(error)  

            weights -= learning_rate * gradient_w
            bias -= learning_rate * gradient_b

        if epoch % 100 == 0:
            y_cap = np.dot(x, weights) + bias
            loss = mse(y, y_cap)
            print(f"Epoch {epoch}: Mean Squared Error = {loss:.4f}")
            print(f"Weights: {weights}")
            print(f"Bias: {bias}")

    return weights, bias


x = np.random.randint(1, 10, size=(100, 2))  
y = np.random.randint(1, 10, size=(100, 1))  

x = (x - np.mean(x, axis=0)) / np.std(x, axis=0)

weights, bias = mini_batch_grad_descent(x, y)

print(f"Final Weights: {weights}")
print(f"Final Bias: {bias}")
```
